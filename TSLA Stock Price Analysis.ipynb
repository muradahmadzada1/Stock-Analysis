```python
data['%-change'] = data['Close'].pct_change()
```

```python
data.head()
```

```python
data['Log returns'] = np.log(data['Close']/data['Close'].shift())
```

```python
data.head()
```

```python
data['Log returns'].std()
```

```python
volatility = data['Log returns'].std()*252**.5
```

```python
vol = str(round(volatility, 5)*100)
print(vol)
```

```python
fig, ax = plt.subplots()
data['Log returns'].hist(ax=ax, bins=50, alpha=0.8, color='r')
ax.set_xlabel("Log return")
ax.set_ylabel("Frequency of log return")
ax.set_title("TSLA volatility: " + vol + "%")
```

```python
data['Moving_Average'] = data['Close'].rolling(5).mean()
```

```python
data.tail()
```

```python
data['Exponential_Moving_Average'] = data['Close'].ewm(span=5, adjust=False).mean()
```

```python
ig, ax = plt.subplots()
data[['Moving_Average', 'Exponential_Moving_Average']].loc['2022-2-18':].plot(ax=ax)
data['Close'].loc['2022-2-18':].plot(ax=ax, alpha=0.5)
```

```python
data.tail()
```

```python
exp1 = data['Close'].ewm(span=12, adjust=False).mean()
exp2 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = exp1 - exp2
data['Signal line'] = data['MACD'].ewm(span=9, adjust=False).mean()
```

```python
data.head()
```

```python
fig, ax = plt.subplots()
data[['MACD', 'Signal line']].plot(ax=ax)
data['Close'].plot(ax=ax, alpha=0.25, secondary_y=True)
```

```python
high30 = data['High'].rolling(30).max()
low30 = data['Low'].rolling(30).min()
data['%K'] = (data['Close'] - low30)*100/(high30 - low30)
data['moving_average_K'] = data['%K'].rolling(3).mean()
```

30-high: Maximum of last 30 trading days

30-low: Minimum of last 30 trading days

%K: (Last close - 30-low)*100 / (30-high - 30-low)

moving_average_K: Simple Moving Average of %K

```python
data.tail()
```

```python
fig, ax = plt.subplots()
data[['%K', 'moving_average_K']].loc['2022-2-18':].plot(ax=ax)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
data['Close'].loc['2022-2-18':].plot(ax=ax, alpha=0.3, secondary_y=True)
     
```
